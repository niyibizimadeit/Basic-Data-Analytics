```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create figure
fig, ax = plt.subplots(figsize=(10, 8))
ax.axis("off")

# Helper function to draw boxes and diamonds
def draw_box(ax, text, xy, boxstyle="round,pad=0.3", fc="white"):
    box = mpatches.FancyBboxPatch(xy, 3, 1.2, boxstyle=boxstyle, fc=fc, ec="black")
    ax.add_patch(box)
    ax.text(xy[0] + 1.5, xy[1] + 0.6, text, ha="center", va="center", wrap=True)
    return box

def draw_diamond(ax, text, xy):
    diamond = mpatches.RegularPolygon((xy[0] + 1.5, xy[1] + 0.6), numVertices=4, radius=1.1, orientation=0.785, fc="white", ec="black")
    ax.add_patch(diamond)
    ax.text(xy[0] + 1.5, xy[1] + 0.6, text, ha="center", va="center", wrap=True)
    return diamond

# Draw flowchart elements
start = draw_box(ax, "Start", (4, 13), boxstyle="round,pad=0.3", fc="white")
cond1 = draw_diamond(ax, "Message valid?\nAND\nType correct?", (4, 11))
decode = draw_box(ax, "Decode message content", (4, 9))
cond2 = draw_diamond(ax, "Message content correct?", (4, 7))
store1 = draw_box(ax, "Store message in database", (4, 5))
error1 = draw_box(ax, "Handle error message", (8, 5))
store2 = draw_box(ax, "Store message in database", (8, 3))
error2 = draw_box(ax, "Handle error message", (0, 9))
store3 = draw_box(ax, "Store message in database", (0, 7))
end = draw_box(ax, "End", (4, 1), boxstyle="round,pad=0.3", fc="white")

# Draw arrows
def draw_arrow(ax, start, end, text=None, offset=(0,0)):
    ax.annotate("", xy=end, xytext=start,
                arrowprops=dict(arrowstyle="->", lw=1.2, color="black"))
    if text:
        ax.text((start[0]+end[0])/2 + offset[0], (start[1]+end[1])/2 + offset[1], text, 
                ha="center", va="center", fontsize=9)

draw_arrow(ax, (5.5, 13), (5.5, 12.2))  # Start to cond1
draw_arrow(ax, (5.5, 10.4), (5.5, 9.8), "True")  # cond1 true
draw_arrow(ax, (5.5, 8.4), (5.5, 7.8))  # decode to cond2
draw_arrow(ax, (5.5, 6.4), (5.5, 5.8), "True")  # cond2 true -> store1
draw_arrow(ax, (5.5, 4.4), (5.5, 2.2))  # store1 -> end

# False from cond2 -> error1
draw_arrow(ax, (6.6, 7.2), (8, 6.2), "False", offset=(0.5,0.3))
draw_arrow(ax, (9.5, 4.4), (5.5, 2.2))  # store2 -> end

# False from cond1 -> error2
draw_arrow(ax, (4.4, 11.2), (1.5, 10.2), "False", offset=(0,0.3))
draw_arrow(ax, (1.5, 8.4), (1.5, 7.8))  # error2 -> store3
draw_arrow(ax, (1.5, 6.4), (5.5, 2.2))  # store3 -> end

# error1 -> store2
draw_arrow(ax, (9.5, 4.4), (9.5, 3.8))

plt.show()
```
