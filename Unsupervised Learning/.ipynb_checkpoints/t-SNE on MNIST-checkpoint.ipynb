```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import random

#Fetching MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=False)

X = mnist.data.astype(np.float32)
y = mnist.target.astype(int)

#standardizing(Optional)
scaler = StandardScaler()
X = scaler.fit_transform(X)

#Using a few samples for speed
n_samples = 2000
rng = np.random.RandomState(42)
idx = rng.choice(len(X), n_samples, replace=True)
X_sub = X[idx]
y_sub = y[idx]

#Using PCA to reduce dimensions to 50
pca2 = PCA(n_components=50, random_state=42)
X_pca50 = pca2.fit_transform(X_sub)

#t-SNE on the PCA reduced data
tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, n_iter=1000, random_state=42, init='pca')
X_tsne = tsne.fit_transform(X_pca50)

# Plotting colored by digit
plt.figure(figsize=(8,8))
scatter = plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y_sub, cmap='tab10', s=8)
plt.legend(*scatter.legend_elements(num=10), title="digit", loc='best')
plt.title("t-SNE on MNIST (subset, PCA->50 pre-reduction)")
plt.xlabel("t-SNE dim1")
plt.ylabel("t-SNE dim2")
plt.show()
```
